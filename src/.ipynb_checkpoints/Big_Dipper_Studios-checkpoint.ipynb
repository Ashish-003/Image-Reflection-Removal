```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
# import matlab.engine
```

```python
ghost_gray_image = cv2.cvtColor(cv2.imread("../images/ghost3.png"), cv2.COLOR_BGR2GRAY)
plt.imshow(ghost_gray_image, cmap="gray")
plt.title("Ghost Gray Image")
plt.show()
```

```python
template = ghost_gray_image[370:490, 630:750]
plt.imshow(template, cmap="gray")
plt.title("User Input Patch")
plt.show()
```

```python
def reflection_removal(gray_img, endpoints):
    return gray_img
```

```python
def templateMatching(ghost_gray_image,template):
  result = cv2.matchTemplate(ghost_gray_image, template,cv2.TM_CCOEFF_NORMED)
  # (minVal, maxVal, minLoc, maxLoc) = cv2.minMaxLoc(result)
  return result
```

```python
reflection_removed_gray_image = reflection_removal(ghost_gray_image, np.array({"x_min": 625, "x_max": 750, "y_min": 400, "y_max": 520}))
plt.imshow(reflection_removed_gray_image, cmap="gray")
plt.title("Reflection Removed Gray Image")
plt.show()
```

```python
threshold = 0.8
img_rgb = np.copy(ghost_gray_image)
w,h = template.shape
result = templateMatching(np.copy(ghost_gray_image),np.copy(template))
loc = np.where( result >= threshold)
for pt in zip(*loc[::-1]):
    cv2.rectangle(img_rgb, pt, (pt[0] + w, pt[1] + h), (0,0,255), 2)
plt.imshow(img_rgb, cmap="gray")
plt.title("Ghost Gray Image")
plt.show()
```

```python
import numpy as np
import scipy
import scipy.ndimage as ndimage
import scipy.ndimage.filters as filters
import matplotlib.pyplot as plt

neighborhood_size = 12
threshold = 0.20

data = result 

data_max = filters.maximum_filter(data, neighborhood_size)
maxima = (data == data_max)
data_min = filters.minimum_filter(data, neighborhood_size)
diff = ((data_max - data_min) > threshold)
maxima[diff == 0] = 0

labeled, num_objects = ndimage.label(maxima)
slices = ndimage.find_objects(labeled)
x, y = [], []
for ey,ex in slices:
    x_center = (ex.start + ex.stop - 1)/2
    x.append(x_center)
    y_center = (ey.start + ey.stop - 1)/2    
    y.append(y_center)

plt.imshow(data, cmap="gray")
# plt.savefig('/tmp/data.png', bbox_inches = 'tight')

plt.autoscale(False)
plt.plot(x,y, 'ro')
print(x)
print(y)
# plt.savefig('/tmp/result.png', bbox_inches = 'tight')
```
